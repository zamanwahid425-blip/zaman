```python
# Step 1: Install dependencies
!pip install reportlab seaborn

# Step 2: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from reportlab.lib.pagesizes import A4
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image
from reportlab.lib.styles import getSampleStyleSheet

# Step 3: Load Dataset
url = "https://raw.githubusercontent.com/dakshtyagi0002/PRODIGY_DS_01/main/population.csv"
data = pd.read_csv(url)

print("‚úÖ Dataset loaded successfully!")
print("Columns in dataset:", data.columns.tolist())
print(data.head())

# Step 4: Identify correct column names
gender_col = None
age_col = None

for col in data.columns:
    if "gender" in col.lower() or "sex" in col.lower():
        gender_col = col
    if "age" in col.lower():
        age_col = col

# Step 5: Create Charts
if gender_col:
    plt.figure(figsize=(6,5))
    sns.countplot(data=data, x=gender_col, palette="Set2")
    plt.title("Gender Distribution")
    plt.savefig("gender_distribution.png")
    plt.close()
else:
    print("‚ö† No Gender column found in dataset!")

if age_col:
    plt.figure(figsize=(8,5))
    sns.histplot(data[age_col], bins=10, kde=True, color='skyblue')
    plt.title("Age Distribution")
    plt.savefig("age_distribution.png")
    plt.close()
else:
    print("‚ö† No Age column found in dataset!")

# Step 6: Generate PDF Report
doc = SimpleDocTemplate("Internship_Task01_Report.pdf", pagesize=A4)
styles = getSampleStyleSheet()
story = []

story.append(Paragraph("Prodigy Infotech Internship", styles["Title"]))
story.append(Spacer(1, 12))
story.append(Paragraph("Task 01: Create distribution charts.", styles["Normal"]))
story.append(Spacer(1, 12))

if gender_col:
    story.append(Paragraph("1. Gender Distribution", styles["Heading2"]))
    story.append(Image("gender_distribution.png", width=400, height=300))
    story.append(Spacer(1, 12))

if age_col:
    story.append(Paragraph("2. Age Distribution", styles["Heading2"]))
    story.append(Image("age_distribution.png", width=400, height=300))
    story.append(Spacer(1, 12))

story.append(Paragraph("üìä Insights:- Shows demographic balance.- Identifies common age range.- Helps analyze population structure.", styles["Normal"]))

doc.build(story)
print("‚úÖ PDF Report Generated!")

# Step 7: Download PDF
from google.colab import files
files.download("Internship_Task01_Report.pdf")
```
